```python
import os
from ultralytics import YOLO

# 1. Load a pre-trained YOLOv8 nano model
model = YOLO("yolov8n.pt")

# 2. Train the model on your custom Arcade dataset
# Ensure the dataset path matches your drive directory
results = model.train(
    data="/content/drive/MyDrive/arcade/arcade.yaml",
    epochs=150,
    imgsz=640,
    batch=16,
    patience=30,
    device=0  # Use GPU
)
```
