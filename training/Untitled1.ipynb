```python
class SegmentationDataset(Dataset):
    def __init__(self, image_dir, mask_dir, transforms=None):
        self.image_paths = sorted(glob(os.path.join(image_dir, "*")))
        self.mask_paths = sorted(glob(os.path.join(mask_dir, "*")))
        assert len(self.image_paths) == len(self.mask_paths), "Images/masks mismatch"
        self.transforms = transforms

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = cv2.imread(self.image_paths[idx], cv2.IMREAD_COLOR)  # BGR
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)

        if mask is None:
            # create empty mask if missing
            mask = np.zeros((img.shape[0], img.shape[1]), dtype=np.uint8)

        # Ensure mask is 0/1 (float)
        mask = (mask > 127).astype("float32")


        if self.transforms:
            augmented = self.transforms(image=img, mask=mask)
            img = augmented["image"]
            mask = augmented["mask"]
        else:
            img = ToTensorV2()(image=img)["image"]
            mask = torch.tensor(mask).unsqueeze(0).float()

        # IMPORTANT: ensure mask shape is (1, H, W)
        if isinstance(mask, torch.Tensor):
            if mask.ndim == 2:  # H x W
                mask = mask.unsqueeze(0)
        else:
            # albumentations sometimes returns numpy array -> convert
            mask = torch.tensor(mask).unsqueeze(0).float()

        return img.float(), mask.float()
```

```python
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
ENCODER = "resnet34"
model = smp.Unet(encoder_name=ENCODER, encoder_weights="imagenet", in_channels=3, classes=1, activation=None)
model = model.to(DEVICE)

# Losses
bce = nn.BCEWithLogitsLoss()

def dice_loss(pred, target, eps=1e-7):
    pred = torch.sigmoid(pred)
    num = 2 * (pred * target).sum(dim=(2,3))
    den = (pred + target).sum(dim=(2,3))
    loss = 1 - (num + eps) / (den + eps)
    return loss.mean()

def combined_loss(pred, target):
    return bce(pred, target) + dice_loss(pred, target)

# optimizer & scheduler
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="max", factor=0.5, patience=3)
```

```python
import numpy as np

@torch.no_grad()
def compute_batch_metrics(pred, target, thr=0.5):
    pred_mask = (torch.sigmoid(pred) > thr).float()
    target = target.float()
    inter = (pred_mask * target).sum(dim=(2,3))
    union = pred_mask.sum(dim=(2,3)) + target.sum(dim=(2,3))
    dice = (2 * inter) / (union + 1e-7)
    # IoU approximate
    iou = inter / ( (pred_mask + target - pred_mask*target).sum(dim=(2,3)) + 1e-7 )
    # return mean across batch
    return float(dice.mean().cpu().numpy()), float(iou.mean().cpu().numpy())
```

```python
# Cell 4: Build model. Prefer segmentation_models_pytorch (SMP) if installed.
def get_model(num_classes=1, encoder_name="resnet34", encoder_weights="imagenet", activation=None):
    if SMP_AVAILABLE:
        if num_classes == 1:
            activation = "sigmoid" if activation is None else activation
        else:
            activation = "softmax2d" if activation is None else activation
        model = smp.Unet(
            encoder_name=encoder_name,
            encoder_weights=encoder_weights,
            in_channels=3,
            classes=num_classes,
            activation=None  # we handle logits -> loss
        )
        return model
    else:
        # Simple U-Net fallback (lightweight). Define a small U-Net here.
        class DoubleConv(nn.Module):
            def __init__(self, in_ch, out_ch):
                super().__init__()
                self.net = nn.Sequential(
                    nn.Conv2d(in_ch, out_ch, 3, padding=1),
                    nn.BatchNorm2d(out_ch),
                    nn.ReLU(inplace=True),
                    nn.Conv2d(out_ch, out_ch, 3, padding=1),
                    nn.BatchNorm2d(out_ch),
                    nn.ReLU(inplace=True)
                )
            def forward(self,x): return self.net(x)

        class SimpleUNet(nn.Module):
            def __init__(self, num_classes):
                super().__init__()
                self.enc1 = DoubleConv(3, 64)
                self.pool = nn.MaxPool2d(2)
                self.enc2 = DoubleConv(64, 128)
                self.enc3 = DoubleConv(128, 256)
                self.bottleneck = DoubleConv(256, 512)
                self.up3 = nn.ConvTranspose2d(512, 256, 2, stride=2)
                self.dec3 = DoubleConv(512, 256)
                self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
                self.dec2 = DoubleConv(256, 128)
                self.up1 = nn.ConvTranspose2d(128, 64, 2, stride=2)
                self.dec1 = DoubleConv(128, 64)
                self.outc = nn.Conv2d(64, num_classes, 1)
            def forward(self, x):
                e1 = self.enc1(x)
                e2 = self.enc2(self.pool(e1))
                e3 = self.enc3(self.pool(e2))
                b = self.bottleneck(self.pool(e3))
                d3 = self.dec3(torch.cat([self.up3(b), e3], dim=1))
                d2 = self.dec2(torch.cat([self.up2(d3), e2], dim=1))
                d1 = self.dec1(torch.cat([self.up1(d2), e1], dim=1))
                out = self.outc(d1)
                return out
        return SimpleUNet(num_classes=num_classes)

model = get_model(num_classes=CONFIG['num_classes'], encoder_name=CONFIG['encoder'], encoder_weights=CONFIG['encoder_weights'])
model.to(CONFIG['device'])
print(model)
```

```python
# Cell 5: loss (BCE + Dice for binary) or CrossEntropy + Dice for multi-class
# Dice implementation (per-class)
def dice_loss_logits(pred, target, eps=1e-6):
    # pred: logits or probabilities? Expect logits not activated; we will apply sigmoid/softmax outside
    if pred.shape[1] == 1:
        pred_probs = torch.sigmoid(pred)
        target = target.float()
        intersection = (pred_probs * target).sum(dim=(2,3))
        union = pred_probs.sum(dim=(2,3)) + target.sum(dim=(2,3))
        dice = (2. * intersection + eps) / (union + eps)
        return 1.0 - dice.mean()
    else:
        pred_probs = torch.softmax(pred, dim=1)
        # target: one-hot [B,C,H,W]
        intersection = (pred_probs * target).sum(dim=(2,3))
        union = pred_probs.sum(dim=(2,3)) + target.sum(dim=(2,3))
        dice = (2. * intersection + eps) / (union + eps)
        return 1.0 - dice.mean()

# Combined loss
def get_loss_fn(num_classes=1, bce_weight=1.0, dice_weight=1.0):
    if num_classes == 1:
        bce = nn.BCEWithLogitsLoss()
        def loss_fn(preds, targets):
            l1 = bce(preds.squeeze(1), targets.squeeze(1))
            l2 = dice_loss_logits(preds, targets)
            return bce_weight*l1 + dice_weight*l2
    else:
        ce = nn.CrossEntropyLoss()
        def loss_fn(preds, targets):
            # preds: [B,C,H,W], targets: one-hot [B,C,H,W] -> convert targets to class indices
            tgt_idx = targets.argmax(dim=1).long()
            l1 = ce(preds, tgt_idx)
            l2 = dice_loss_logits(preds, targets)
            return bce_weight*l1 + dice_weight*l2
    return loss_fn

loss_fn = get_loss_fn(CONFIG['num_classes'])
```

```python
# Cell 6: Create dataloaders (make sure folders exist)
train_ds = ArteryDataset(CONFIG['train_images'], CONFIG['train_masks'], img_size=CONFIG['img_size'], augment=True, num_classes=CONFIG['num_classes'])
val_ds   = ArteryDataset(CONFIG['val_images'], CONFIG['val_masks'], img_size=CONFIG['img_size'], augment=False, num_classes=CONFIG['num_classes'])

train_loader = DataLoader(train_ds, batch_size=CONFIG['batch_size'], shuffle=True, num_workers=CONFIG['num_workers'], pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=CONFIG['batch_size'], shuffle=False, num_workers=CONFIG['num_workers'], pin_memory=True)

print("Train size:", len(train_ds), "Val size:", len(val_ds))
```

```python
optimizer = AdamW(model.parameters(), lr=CONFIG['learning_rate'], weight_decay=CONFIG['weight_decay'])
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=3, factor=0.5)
scaler = GradScaler()  # for mixed precision
```

```python
# Cell 8: metrics + save checkpoint helpers
import math
def iou_score(preds, targets, threshold=0.5, eps=1e-6):
    # preds: logits [B,1,H,W] or [B,C,H,W]; targets: same shape
    if preds.shape[1] == 1:
        probs = torch.sigmoid(preds)
        preds_bin = (probs > threshold).float()
        targets_bin = targets
        inter = (preds_bin * targets_bin).sum(dim=(2,3))
        union = preds_bin.sum(dim=(2,3)) + targets_bin.sum(dim=(2,3)) - inter
        iou = ((inter + eps) / (union + eps)).mean().item()
        return iou
    else:
        probs = torch.softmax(preds, dim=1)
        preds_cls = probs.argmax(dim=1)
        target_cls = targets.argmax(dim=1)
        inter = ((preds_cls==target_cls) & (target_cls>0)).float().sum()
        union = ((preds_cls>0) | (target_cls>0)).float().sum()
        return ((inter+eps)/(union+eps)).item()

def save_checkpoint(epoch, model, optimizer, path):
    state = {
        'epoch': epoch,
        'model_state': model.state_dict(),
        'optim_state': optimizer.state_dict()
    }
    torch.save(state, path)
    print(f"Saved checkpoint: {path}")
```

```python
# Cell 9: Training loop
best_val_iou = -1.0
best_epoch = -1
patience_counter = 0

for epoch in range(1, CONFIG['epochs']+1):
    model.train()
    train_loss = 0.0
    train_steps = 0
    pbar = tqdm(train_loader, desc=f"Epoch {epoch} training", leave=False)
    for imgs, masks in pbar:
        imgs = imgs.to(CONFIG['device'], non_blocking=True).float()
        masks = masks.to(CONFIG['device'], non_blocking=True).float()

        optimizer.zero_grad()
        with autocast():
            outputs = model(imgs)
            loss = loss_fn(outputs, masks)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()
        train_steps += 1
        pbar.set_postfix({'loss': f"{train_loss/train_steps:.4f}"})

    avg_train_loss = train_loss / max(1, train_steps)

    # Validation
    model.eval()
    val_loss = 0.0
    val_steps = 0
    val_iou = 0.0
    with torch.no_grad():
        for imgs, masks in tqdm(val_loader, desc=f"Epoch {epoch} validating", leave=False):
            imgs = imgs.to(CONFIG['device']).float()
            masks = masks.to(CONFIG['device']).float()
            outputs = model(imgs)
            loss = loss_fn(outputs, masks)
            val_loss += loss.item()
            val_steps += 1
            val_iou += iou_score(outputs.detach().cpu(), masks.detach().cpu())

    avg_val_loss = val_loss / max(1, val_steps)
    avg_val_iou = val_iou / max(1, val_steps)

    print(f"Epoch {epoch} | Train loss: {avg_train_loss:.4f} | Val loss: {avg_val_loss:.4f} | Val IOU: {avg_val_iou:.4f}")

    # scheduler step (reduce lr on plateau)
    scheduler.step(avg_val_iou)

    # checkpoint best
    if avg_val_iou > best_val_iou:
        best_val_iou = avg_val_iou
        best_epoch = epoch
        save_checkpoint(epoch, model, optimizer, os.path.join(CONFIG['save_dir'], "best_model.pth"))
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= CONFIG['patience']:
            print("Early stopping triggered.")
            break

print(f"Training finished. Best val IOU: {best_val_iou:.4f} at epoch {best_epoch}")
```
